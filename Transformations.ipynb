Importamos paquetes y cargamos el csv

```python
import pandas as pd
import numpy as np
import json
import re
```

```python
df = pd.read_csv(r'movies_dataset.csv')
```

FUNCIONES UTILIZADAS PARA DESARROLLAR LAS TRANSFORMACIONES

```python
def convertir_lista_a_cadena(x):
    if isinstance(x, list):
        return ', '.join(x)
    else:
        return x
```

```python
#Se utilizo para desanidar las columnas
def extraer_datos(df, columna):
    def extraer_valores(valor):
        if isinstance(valor,str):
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, valor)
            return coincidencias
        elif isinstance(valor, list):
            valores = []
            for elemento in valor:
                if 'name' in elemento:
                    valores.append(elemento['name'])
                    return valores
        elif isinstance(valor, dict):
            if 'name' in valor:
                return valor['name']
        return None
    df[columna] = df[columna].apply(extraer_valores)

    return df
```

Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.

```python
df['revenue'].fillna(0, inplace=True)
df['budget'].fillna(0, inplace=True)
```

Los valores nulos del campo release date deben eliminarse.

```python
df.dropna(subset=['release_date'], inplace=True)
```

De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
df['release_year'] = df['release_date'].dt.year.fillna(0).astype(int)
```

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['return'] = df['revenue'] / df['budget']
df['return'].fillna(0, inplace=True)
df['return'] = df['return'].replace([np.inf, -np.inf], 0)
```

Eliminar las columnas que no serán utilizadas, video,imdb_id,adult,original_title,vote_count,poster_path y homepage

```python
df.drop(['video', 'imdb_id', 'adult', 'original_title', 'vote_count', 'poster_path', 'homepage'], axis=1, inplace=True)
```

Estas transformaciones se usaron para las funciones de la api

```python
df['release_date'] = pd.to_datetime(df['release_date'])
    
# Crear una nueva columna 'mes' con el nombre del mes de la fecha de lanzamiento
df['release_month'] = df['release_date'].dt.month_name()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
    
# Crear una nueva columna 'dia' con el día de la fecha de lanzamiento
df['release_day'] = df['release_date'].dt.day_name()
```

Dataset final para el consumo de la API

```python
df.to_csv('movies_dataset_final.csv', index=False)
```
