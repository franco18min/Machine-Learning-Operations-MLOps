FUNCIONES UTILIZADAS PARA DESARROLLAR LAS TRANSFORMACIONES

```python
def convertir_lista_a_cadena(x):
    if isinstance(x, list):
        return ', '.join(x)
    else:
        return x
```

```python
#Se utilizo para desanidar las columnas
def extraer_datos(df, columna):
    def extraer_valores(valor):
        if isinstance(valor,str):
            pattern = r"'name': '([^']*)'"
            coincidencias = re.findall(pattern, valor)
            return coincidencias
        elif isinstance(valor, list):
            valores = []
            for elemento in valor:
                if 'name' in elemento:
                    valores.append(elemento['name'])
                    return valores
        elif isinstance(valor, dict):
            if 'name' in valor:
                return valor['name']
        return None
    df[columna] = df[columna].apply(extraer_valores)

    return df
```

Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.

```python
df['revenue'].fillna(0, inplace=True)
df['budget'].fillna(0, inplace=True)
```

Los valores nulos del campo release date deben eliminarse.

```python
df.dropna(subset=['release_date'], inplace=True)
```

De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
df['release_year'] = df['release_date'].dt.year.fillna(0).astype(int)
```

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['return'] = df['revenue'] / df['budget']
df['return'].fillna(0, inplace=True)
df['return'] = df['return'].replace([np.inf, -np.inf], 0)
```

Elimino de la columna 'crews' el caracter "\\" y transformo la columna en lista de diccionarios ya que inicialmente esta en formato string

```python
def convert_str_to_list(column_str):
    column_str = column_str.replace('\\', '')
    return ast.literal_eval(column_str)

df_credits['crew'] = df_credits['crew'].apply(convert_str_to_list)
```

Ya convertido los elementos extraigo el nombre de aquellas personas que son Directores de las peliculas en una nueva columna

```python
def get_director_name(crew):
    for member in crew:
        if member['job'].lower() == 'director':
            return member['name']
    return None

df_credits['director'] = df_credits['crew'].apply(get_director_name)
```
